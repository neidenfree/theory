```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.io.wavfile import write
from matplotlib.pyplot import figure
```

```python
pi2 = np.pi * 2
antsec = 1
fs = 44100
x = np.linspace(0, antsec, fs)
y = 0.3*np.sin(pi2 * x * 55) + 0.1*np.sin(pi2 * x * 440)
plt.plot(x, y)
plt.grid()
plt.show()
write('file.wav', fs, y)
```

```python
a1 = 1
a2 = 1
n1 = 1
n2 = 10
t = np.linspace(0, 1, 1000)

f1 = a1 * np.sin(pi2 * t * n1)
f2 = a2 * np.sin(pi2 * t * n2)
f = f1 + f2

figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

# plt.subplot(3, 1, 1)
plt.plot(t, f1)
plt.grid()

# plt.subplot(3, 1, 2)
plt.plot(t, f2)
plt.grid()

# plt.subplot(3, 1, 3)
plt.plot(t, f)
plt.grid()

plt.show()
```
