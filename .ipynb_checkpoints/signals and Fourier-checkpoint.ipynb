```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.io.wavfile import write
from matplotlib.pyplot import figure
import plotly.express as px
import wavio
import winsound
```

```python
def add_noise(signal, intensity):
    rand = np.random.randint(intensity, size=signal.shape) / 20
    return signal + rand

def plotly_line(x, y):
    fig = px.line(x=x, y=y)
    fig.show(config={
        'scrollZoom':True
        
    })
    
```

```python
pi2 = np.pi * 2
antsec = 2
fs = 1000
x = np.linspace(0, antsec, fs)
y = np.sin(pi2 * x * 1)
plt.plot(x, add_noise(y, 100))
plt.grid()
plt.show()

```

```python
a1 = 1
a2 = 0.3
n1 = 55
n2 = 220

fs = 44100
sec = 1
t = np.linspace(0, sec, int(fs*sec))

f1 = a1 * np.sin(pi2 * t * n1)
f2 = a2 * np.sin(pi2 * t * n2)
f = f1 + f2

figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

# plt.plot(t, f)
# plt.grid()
# plt.show()

wavio.write('file.wav', f, fs, sampwidth=4)
winsound.PlaySound('file.wav', winsound.SND_FILENAME)
```

# Square and triangle wave


```python
##### Combination of different waves
fs = 44100
sec = 1
T = 440

period = np.hstack((np.ones(T//2), -np.ones(T//2)))
y_square = np.tile(period, int(fs*sec//T))

period = [4 * x / T - 1 for x in range(0, T//2)] + [3 - 4*x/T for x in range(T//2, T)]
y_triangle = np.tile(period, int(fs*sec//T))

period = [4 * x / T - 1 for x in range(0, T)]
y_saw = np.tile(period, int(fs*sec//T))


x = np.linspace(0, sec, int(sec*fs))
y_sin = np.sin(pi2 * x * T)

y = y_triangle

a = [55]
[a.append(a[len(a)-1]*2) for _ in range(1,5)] 
y = sum([np.sin(pi2 * x * i) for i in [2*j for j in a]])

a = [32]
[a.append(a[len(a)-1]*2) for _ in range(3,6)] 
y += sum([np.sin(pi2 * x * i) for i in [2*j for j in a]])

a = [41.2]
# a = [60]
[a.append(a[len(a)-1]*2) for _ in range(4,5)] 
y += sum([np.sin(pi2 * x * i) for i in [2*j for j in a]])
# plotly_line(x[:6000], y[:6000])

wavio.write('file.wav', y, fs, sampwidth=2)
winsound.PlaySound('file.wav', winsound.SND_FILENAME)
```

# Fourier Transform


```python
four = np.fft.fft(y)
four = four[:len(four)]
xaxis = np.linspace(0, fs, int(fs*sec))
plotly_line(xaxis[:len(xaxis)//2], 2/len(four) * abs(four)[:len(four)//2])
print(four.shape, xaxis.shape)
```

```python
# I want to remove all frequencies above 1000 Hz from signal using Fourier transform. First, I need to get my DFT of signal.

def sound_filter(signal, fs, sec, filter_type, freq):
    four = np.fft.fft(signal)
    xaxis = np.linspace(0, fs, int(fs*sec))
    i = 0
    while True:
        i += 1
        if xaxis[i] > freq:
            break
    if filter_type=='lowpass':
        four[i:] = 0
    elif filter_type == 'highpass':
        four[:i] = 0
        four[-i:] = 0
    print(np.count_nonzero(four==0))
    plt.plot(abs(four))
    return np.fft.ifft(four)

# sound = sound_filter(y, fs, sec, filter_type='lowpass', freq=300)
sound = sound_filter(y, fs, sec, filter_type='highpass', freq=300)

wavio.write('file.wav', sound, fs, sampwidth=2)
winsound.PlaySound('file.wav', winsound.SND_FILENAME)
```
