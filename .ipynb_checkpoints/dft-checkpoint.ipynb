```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
from matplotlib.pyplot import figure
figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
# y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + 1.5*np.sin(200*2*np.pi*x)
y = np.sin(100 * 2 *np.pi*x)
yf = scipy.fftpack.fft(y)

xf = np.linspace(0.0, 1.0/(2.0*T), round(N/2))

plt.subplot(3, 1, 1)
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.grid()
plt.subplot(3, 1, 2)
plt.plot(y)
plt.grid()
plt.show()

# plt.subplot(3, 1, 3)
# plt.plot(dft_s)
# plt.grid()
# plt.show()
```

$$X_k = \sum_{j=0}^{n-1} x_j e^{-2\pi i k j / n}$$

```python
def DFT_slow(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)

```

```python
figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')


n = 1000
j = np.arange(n)
k = 20

cs = np.sin(-2* np.pi * j * k / n)

plt.plot(cs)
plt.grid()
```
